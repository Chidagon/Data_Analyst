#### COMENZAMOS CON EL PLANTEAMIENTO DE HIPÓTESIS.

La primera hipótesis es comprobar si el porcentaje de hombres en Madrid, Andalucia y Extremadura tienen algún tipo de diferencia.

Planteamiento de hipótesis: 

- `H0`: `no hay diferencia` en el porcentaje de hombres de Madird,Andalucía y Extremadura
- H1: hay diferencia en el porcentaje de hombres de las 3 comunidases

```python
from scipy.stats import f_oneway
```

```python
alpha=0.05
sample1 =  MADRID['porcentaje_hombres']
sample2 =  ANDALUCIA['porcentaje_hombres']
simple3 =  EXTREMADURA['porcentaje_hombres']
```

```python
f, p = f_oneway(sample1, sample2, simple3)
print("One-way ANOVA")
print("F-value:", f)
print("p-value:", p)
```

```python
if p < alpha:
    print("Rechazar la hipótesis nula:  no hay diferenciaen el porcentaje de hombres de Madird,Andalucía y Extremadura")
else:
    print("No se puede rechazar la hipótesis nula: hay diferencia significativa en el porcentaje de hombres de Madird,Andalucía y Extremadura")
```

Como podemos comprobar mediante ANOVA no hay diferencia en el porcentaje de hombes de estas comunidades
