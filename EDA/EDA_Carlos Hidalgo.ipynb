```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.display.max_columns=None
```

```python
df=pd.read_csv(r"C:\Users\DEEPGAMING\Downloads\Municipios_de_Espa%C3%B1a._Padron2017.csv")
df
```

```python
a = df[["Texto","Cod_CCAA","PAD_2C02","PAD_2C21","PAD_2C40","PAD_3C03","PAD_3C04","Shape__Area","Shape__Length"]]
a.rename(columns={'Texto':'Pueblo','Cod_CCAA':'Cod_CCAA', 'PAD_2C02':'Población',"PAD_2C21":"Hombres","PAD_2C40":"Mujeres","PAD_3C03":"Españoles","PAD_3C04":"Extranjeros","Shape__Area":"Area","Shape__Length":"Longuitud"}, inplace=True)
a
```

```python
a1=df[["PAD_2C04","PAD_2C05","PAD_2C06","PAD_2C07","PAD_2C08","PAD_2C09","PAD_2C10","PAD_2C11","PAD_2C12","PAD_2C13","PAD_2C14","PAD_2C15","PAD_2C16","PAD_2C17","PAD_2C18","PAD_2C19"]]
a1
```

```python
a1["menores"]= a1["PAD_2C04"]+ a1["PAD_2C05"]+ a1["PAD_2C06"]
a1["edad_laboral"]= a1["PAD_2C07"]+ a1["PAD_2C08"]+ a1["PAD_2C09"]+ a1["PAD_2C10"]+ a1["PAD_2C11"]+ a1["PAD_2C12"]+ a1["PAD_2C13"]+ a1["PAD_2C14"]+ a1["PAD_2C15"]
a1["jubilados"]= a1["PAD_2C16"]+ a1["PAD_2C17"]+ a1["PAD_2C18"]+ a1["PAD_2C19"]
a1
```

```python
a2 = a1[["menores","edad_laboral","jubilados"]]
a2
```

```python
ESPAÑA=pd.concat([a,a2], axis=1)
ESPAÑA.dropna(inplace=True)
```

```python
ESPAÑA.rename(columns={"Shape_Area": "Area", "Shape_Leng": "Longitud"}, inplace=True)
ESPAÑA
```

```python
ESPAÑA["porcentaje_hombres"]=ESPAÑA["Hombres"]/ESPAÑA["Población"]*100
ESPAÑA["porcentaje_mujeres"]=ESPAÑA["Mujeres"]/ESPAÑA["Población"]*100
ESPAÑA["porcenjate_españoles"]=ESPAÑA["Españoles"]/ESPAÑA["Población"]*100
ESPAÑA["poetentaje_extranjeros"]=ESPAÑA["Extranjeros"]/ESPAÑA["Población"]*100
ESPAÑA
```
